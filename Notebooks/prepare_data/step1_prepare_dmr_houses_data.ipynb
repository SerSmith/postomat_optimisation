# Тетрадка с подготовкой данных <code>raw_dmr_houses_data</code>

<code>raw_dmr_houses_data</code> используется в <code>apartment_houses_all_data.ipynb</code> для создания и заливки в БД таблички <code>apartment_houses_all_data</code>

Где взять сырые данные:
- нужно зайти по ссылке или найти поиском на https://data.mos.ru
- найти датасет "Адресный реестр объектов недвижимости города Москвы"
- скачать его в формате json

```python
import os
import json
import pandas as pd
from postamats.utils import prepare_data
from postamats.utils.connections import DB
from postamats.global_constants import RAW_DMR_NAME
from postamats.utils.connections import PATH_TO_ROOT

pd.set_option('display.max_columns', None)
```

```python
# папка с json https://data.mos.ru/opendata/60562/data/table?versionNumber=3&releaseNumber=823
# нужно зайти по ссылке или найти поиском на https://data.mos.ru
# датасет Адресный реестр объектов недвижимости города Москвы
# и скачать его в формате json
DATA_FOLDER = os.path.join(PATH_TO_ROOT, 'data', 'dmr')
DATA_FILE = 'data-29580-2022-10-20.json'

CONFIG_PATH = os.path.join(PATH_TO_ROOT, 'db_config.json')
```

```python
prepared_dmr = prepare_data.prepare_dmr_houses_data(os.path.join(DATA_FOLDER, DATA_FILE), geo=False)
```

```python
prepared_dmr
```

```python
with open(CONFIG_PATH, mode='r') as db_file:
    db_config = json.load(db_file)

database = DB(db_config)
database.load_to_bd(prepared_dmr, RAW_DMR_NAME)
```
